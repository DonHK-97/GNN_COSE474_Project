```python
#I tried to add the scores from someother dataset but since we have sentiment score
#At leat we need some relevancy for that.
import pandas as pd
import numpy as np
import random

scoring = pd.read_csv('scores.csv')
scoring
```

```python
#Make rating data from the Sentiment Score.
#Add feature Ratings.
scoring['Ratings'] = 0.0
labels = scoring['LABEL']
rates = scoring['Ratings']

for i in range(0, len(labels)):
    if labels[i] == 'POSITIVE': rates[i] = round(random.uniform(3,5),1)
    else: rates[i] = round(random.uniform(0,3),1)
#Using round function so that we could see the number with a format of X.X
scoring
```

```python
#Save and terminate for memory space.
scoring.to_csv('scores.csv')
```

```python
del[[scoring, labels, rates]]
```

```python
#Read another dataset to concatanate
product = pd.read_csv('product.csv')
product.head()
```

```python
#Since our whole data doesn't depend on ID's, get rid of them
product = product.drop(['order_id','user_id','page_id','target'], axis=1)
product.head()
```

```python
#Concat those two dataset [ Sentiment Score and Rating & Product Commercial Infos]
final_dataset = pd.concat([scoring, product], axis=1)
final_dataset.head()
```

```python
#Cleaning other useless colmnus and row includes 'NaN'
final_dataset = final_dataset.drop(['Unnamed: 0'], axis=1)
final_dataset = final_dataset.dropna(axis=0)
final_dataset
```

```python
#All done!
final_dataset.to_csv('final_dataset.csv')
```
