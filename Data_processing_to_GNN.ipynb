```python
#I tried to add the scores from someother dataset but since we have sentiment score
#At leat we need some relevancy for that.
import pandas as pd
import numpy as np
import random

scoring = pd.read_csv('text_score.csv')
scoring.head()
```

```python
#Make rating data from the Sentiment Score.
#Add feature Ratings.
scoring['Ratings'] = 0.0
labels = scoring['LABEL']
rates = scoring['Ratings']

for i in range(0, len(labels)):
    if labels[i] == 'POSITIVE': rates[i] = round(random.uniform(3,5),1)
    else: rates[i] = round(random.uniform(0,3),1)
#Using round function so that we could see the number with a format of X.X
scoring
```

```python
scoring.to_csv('scores.csv')
```
